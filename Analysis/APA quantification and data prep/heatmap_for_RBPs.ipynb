```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
rbp_df = pd.read_csv('/data1/APA/Paul_ALS_Data/results/avg_log2FC_results.csv', index_col=0)
rbp_df
```

```python
# remove columns with Endothelial in their name
rbp_df = rbp_df[rbp_df.columns.drop(list(rbp_df.filter(regex='Endothelial')))]
```

```python
# plot clustered heatmap for rbp_df
sns.clustermap(rbp_df, cmap='RdBu_r', figsize=(20, 20), vmin=-1, vmax=1)
```
