```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
os.chdir('/data1/APA/Paul_ALS_Data/bams_in/subscelltype_bamfiles/Mapper_outs/')
```

```python
def plot_redu(df, name):
    dat = df['REDu']
    dat.replace([np.inf, -np.inf], np.nan, inplace=True)
    dat.dropna(inplace=True)
    ax = sns.distplot(dat)
    ax.set(xlabel='REDu', ylabel='Density')
    plt.axvline(dat.median(),
                color='red', label=round(dat.median(),2))
    plt.legend()
    outname = "REDU_plots/" + name + '_REDu_distplots.png'
    plt.savefig(outname)
    plt.clf()
```

```python
for ct_cn in CTs:
    file = ct_cn + "/gene.txt"
    df = pd.read_csv(file, sep='\t')
    print(ct_cn)
    print(df.shape)
    plot_redu(df, ct_cn)
```

```python
# lets read in a maaper outout for Excitatory neurons and see what we have
df = pd.read_csv('Excitatory_C9ALSvsCTRL/gene.txt', sep='\t')
df.head()
```

```python
# lets filter for REDu > 0.5 and REDu.pval < 0.05
long_df = df[(df['REDu'] < -1) & (df['REDu.pval'] < 0.05)] # this genes that are longer in ALS than CTRL
short_df = df[(df['REDu'] > 1) & (df['REDu.pval'] < 0.05)] # this genes that are shorter in ALS than CTLR
# sort by REDu descending
long_df.sort_values(by=['REDu'], ascending=True, inplace=True)
short_df.sort_values(by=['REDu'], ascending=False, inplace=True)
# lets see what we have
long_df.head()
```

```python
# ok lets do all this for all the cell types and C9ALSvsCTRL and sALSvsCTRL
main_cts = ['Excitatory', 'Inhibitory', 'Astrocytes', 'Oligodendrocytes', 'Microglia', 'OPC']
CTs = [ct for ct in os.listdir() if 'CTRL' in ct and any(main_ct in ct for main_ct in main_cts)]
CTs = list(set(CTs))
CTs
```

```python
# ok lets do all this for all the cell types and C9ALSvsCTRL and sALSvsCTRL
for CT in CTs:
    print(CT)
    df = pd.read_csv(f'{CT}/gene.txt', sep='\t')
    long_df = df[(df['REDu'] < -1) & (df['REDu.pval'] < 0.001)] # this genes that are longer in ALS than CTRL
    short_df = df[(df['REDu'] > 1) & (df['REDu.pval'] < 0.001)] # this genes that are shorter in ALS than CTLR
    long_df['REDu'] = long_df['REDu'] * -1
    # sort by REDu descending
    long_df.sort_values(by=['REDu'], ascending=False, inplace=True)
    short_df.sort_values(by=['REDu'], ascending=False, inplace=True)
    # remove inf and -inf and dropna
    long_df.replace([np.inf, -np.inf], np.nan, inplace=True)
    long_df.dropna(inplace=True)
    short_df.replace([np.inf, -np.inf], np.nan, inplace=True)
    short_df.dropna(inplace=True)
    # write the gene and REDu values to a file
    long_df[['gene', 'REDu']].to_csv(f'results/genes_for_pathways/{CT}_lengthend_genes.txt', sep='\t', index=False, header=False)
    short_df[['gene', 'REDu']].to_csv(f'results/genes_for_pathways/{CT}_shorend_genes.txt', sep='\t', index=False, header=False)


    
```
